```python
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    """

    if conversion_type == "1":  # meters to feet
        return value * 3.28084

    elif conversion_type == "2":  # feet to meters
        return value / 3.28084

    elif conversion_type == "3":  # kg to pounds
        return value * 2.20462

    elif conversion_type == "4":  # pounds to kg
        return value / 2.20462

    elif conversion_type == "5":  # liters to gallons
        return value * 0.264172

    elif conversion_type == "6":  # gallons to liters
        return value / 0.264172

    else:
        return None


def unit_converter():
    """Main function that interacts with the user and performs unit conversion."""

    print("Select the conversion you want to perform:\n")

    print("1. Convert Meters (m) → Feet (ft)")
    print("2. Convert Feet (ft) → Meters (m)")
    print("3. Convert Kilograms (kg) → Pounds (lbs)")
    print("4. Convert Pounds (lbs) → Kilograms (kg)")
    print("5. Convert Liters (L) → Gallons (gal)")
    print("6. Convert Gallons (gal) → Liters (L)\n")

    conversion_type = input("Enter the number corresponding to your choice (1–6): ")

    try:
        value = float(input("Enter the value you want to convert: "))

        result = convert_units(conversion_type, value)

        if result is None:
            print("Error: Invalid conversion option selected.")
        else:
            print("Converted value =", result)

    except ValueError:
        print("Error: Please enter a valid numeric value.")


unit_converter()
```

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Finds the maximum value in a list of numbers.

    """
    return max(numbers)


def find_min(numbers):
    """
    Finds the minimum value in a list of numbers.

    """
    return min(numbers)


def main():
    """Main function that asks the user for operation choice and numbers."""

    print("Choose the operation you want to perform:\n")

    print("1. Find the Sum of the numbers")
    print("2. Find the Average of the numbers")
    print("3. Find the Maximum number")
    print("4. Find the Minimum number\n")

    choice = input("Enter the number corresponding to your choice (1–4): ")

    try:
        numbers_input = input(
            "Enter a list of numbers separated by spaces (example: 10 20 30 40): "
        )

        numbers = [float(num) for num in numbers_input.split()]

        if len(numbers) == 0:
            print("Error: The list of numbers cannot be empty.")
            return

        if choice == "1":
            result = calculate_sum(numbers)
            print("The sum of the numbers is:", result)

        elif choice == "2":
            result = calculate_average(numbers)
            print("The average of the numbers is:", result)

        elif choice == "3":
            result = find_max(numbers)
            print("The maximum number is:", result)

        elif choice == "4":
            result = find_min(numbers)
            print("The minimum number is:", result)

        else:
            print("Error: Invalid operation selected.")

    except ValueError:
        print("Error: Please enter valid numeric values only.")


main()
```

Exercise on List Manipulation:

```python
#1. Extract Every Other Element:
def extract_every_other(lst):
    """Return every other element from the list, starting with the first element."""
    return lst[::2]

numbers = [3, 4, 5, 6, 7, 8]

result = extract_every_other(numbers)

print("Original list:", numbers)
print("Every other element:", result)
```

```python
#2. Slice a Sublist:
def get_sublist(lst, start, end):
    """
    This function returns a part of the list from start index to end index.
    """
    return lst[start:end+1]

numbers = [3, 4, 5, 6, 7, 8]
print(get_sublist(numbers, 1, 4))
```

```python
#3. Reverse a List Using Slicing:
def reverse_list(lst):
    """
    This function reverses the list using slicing.
    """
    return lst[::-1]

numbers = [3, 4, 5, 6, 7, 8]
print(reverse_list(numbers))
```

```python
#4. Remove the First and Last Elements:
def remove_first_last(lst):
    """
    This function removes the first and last elements of a list.
    """
    return lst[1:-1]

numbers = [3, 4, 5, 6, 7, 8]
print(remove_first_last(numbers))
```

```python
#5. Get the First n Elements:
def get_first_n(lst, n):
    """
    This function returns the first n elements of a list.
    """
    return lst[:n]

numbers = [3, 4, 5, 6, 7, 8]
print(get_first_n(numbers, 2))
```

```python
#6. Extract Elements from the End:
def get_last_n(lst, n):
    """
    This function returns the last n elements of a list.
    """
    return lst[-n:]

numbers = [3, 4, 5, 6, 7, 8]
print(get_last_n(numbers, 3))
```

```python
#7. Extract Elements in Reverse Order:
def reverse_skip(lst):
    """
    This function returns every second element from the list
    starting from the second last element going backwards.
    """
    return lst[-2::-2]

numbers = [3, 4, 5, 6, 7, 8]
print(reverse_skip(numbers))
```

Exercise on Nested List:

```python
# 1. Flatten a Nested List
def flatten(lst):
    """Flatten a nested list into a single list."""
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(item)
        else:
            flat_list.append(item)
    return flat_list

nested_list1 = [[10, 20], [30, 40], [50]]
print("Flattened:", flatten(nested_list1))
```

```python
# 2. Accessing Nested List Elements
def access_nested_element(lst, indices):
    """
    Extracts an element from a nested list using given indices.

    """

    element = lst
    for index in indices:
        element = element[index]
    return element


nested_list2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

result = access_nested_element(nested_list2, indices)

print("The extracted element is:", result)
```

```python
# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list.

    """

    total = 0

    for element in lst:
        if isinstance(element, list):
            total += sum_nested(element)
        else:
            total += element

    return total


nested_list3 = [[1, 2], [3, [4, 6]], 8]

result = sum_nested(nested_list3)

print("The sum of all elements in the nested list is:", result)
```

```python
# 4. Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    """

    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))  # recursive call
        elif item != elem:
            result.append(item)

    return result

nested_list4 = [[1, 2], [3, 2], [4, 5]]
elem = 2
output = remove_element(nested_list4, elem)

print("Modified list:", output)
```

```python
# 5. Find the Maximum Element in a Nested List
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    """

    maximum = None

    for element in lst:
        if isinstance(element, list):
            value = find_max(element)  # recursive call
        else:
            value = element

        if maximum is None or value > maximum:
            maximum = value

    return maximum

nested_list5 = [[1, 2], [3, [4, 5]], 6]

result = find_max(nested_list5)

print("Maximum element in the nested list is:", result)
```

```python
# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    """

    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1

    return count


# Example
nested_list6 = [[1, 5], [5, 3], [2, 3]]
elem = 2

print("Occurrences:", count_occurrences(nested_list6, elem))
```

```python
# 7. Flatten a List of Lists of Lists
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    """

    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result

nested_list7 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

print("Flattened list:", deep_flatten(nested_list7))
```

```python
# 8. Nested List Average
def average_nested(lst):
    """Calculate the average of all elements in a nested list."""
    total_sum = sum_nested(lst)
    def count_elements(l):
        count = 0
        for item in l:
            if isinstance(item, list):
                count += count_elements(item)
            else:
                count += 1
        return count
    total_count = count_elements(lst)
    return total_sum / total_count if total_count != 0 else 0

nested_list8 = [[2, 4], [6, 8], [10, 12]]
print("Average:", average_nested(nested_list8))
```

Experiment: How Fast is Numpy?

```python
import numpy as np
import time
```

1. Element-wise Addition:

```python
def add_lists(list1, list2):
    start = time.time()
    result = [a + b for a, b in zip(list1, list2)]
    end = time.time()
    print("Python list addition time:", end - start, "seconds")
    return result

def add_arrays(arr1, arr2):
    start = time.time()
    result = arr1 + arr2
    end = time.time()
    print("NumPy array addition time:", end - start, "seconds")
    return result
```

2. Element-wise Multiplication

```python
def multiply_lists(list1, list2):
    start = time.time()
    result = [a * b for a, b in zip(list1, list2)]
    end = time.time()
    print("Python list multiplication time:", end - start, "seconds")
    return result

def multiply_arrays(arr1, arr2):
    start = time.time()
    result = arr1 * arr2
    end = time.time()
    print("NumPy array multiplication time:", end - start, "seconds")
    return result
```

3. Dot Product

```python
def dot_lists(list1, list2):
    start = time.time()
    result = sum(a * b for a, b in zip(list1, list2))
    end = time.time()
    print("Python list dot product time:", end - start, "seconds")
    return result

def dot_arrays(arr1, arr2):
    start = time.time()
    result = np.dot(arr1, arr2)
    end = time.time()
    print("NumPy array dot product time:", end - start, "seconds")
    return result
```

4. Matrix Multiplication

```python
def matmul_lists(mat1, mat2):
    size = len(mat1)
    start = time.time()
    result = [[sum(mat1[i][k] * mat2[k][j] for k in range(size))
               for j in range(size)] for i in range(size)]
    end = time.time()
    print("Python list matrix multiplication time:", end - start, "seconds")
    return result

def matmul_arrays(mat1, mat2):
    start = time.time()
    result = np.dot(mat1, mat2)
    end = time.time()
    print("NumPy array matrix multiplication time:", end - start, "seconds")
    return result
```

```python
def run_experiment(list_size=1_000_000, matrix_size=1000):
    print("Element-wise Operations ")

    list1 = list(range(list_size))
    list2 = list(range(list_size))
    arr1 = np.arange(list_size)
    arr2 = np.arange(list_size)

    # Addition
    add_lists(list1, list2)
    add_arrays(arr1, arr2)

    # Multiplication
    multiply_lists(list1, list2)
    multiply_arrays(arr1, arr2)

    # Dot product
    dot_lists(list1, list2)
    dot_arrays(arr1, arr2)

    print("\nMatrix Multiplication")

    mat_list1 = [[i + j for j in range(matrix_size)] for i in range(matrix_size)]
    mat_list2 = [[i + j for j in range(matrix_size)] for i in range(matrix_size)]

    mat_array1 = np.arange(matrix_size * matrix_size).reshape(matrix_size, matrix_size)
    mat_array2 = np.arange(matrix_size * matrix_size).reshape(matrix_size, matrix_size)

    # Matrix multiplication
    matmul_lists(mat_list1, mat_list2)
    matmul_arrays(mat_array1, mat_array2)

run_experiment()
```
